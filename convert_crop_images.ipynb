```python
import cv2
import matplotlib.pyplot as plt
import glob
from natsort import natsorted
```

```python
images = natsorted(glob.glob("*map.pgm"))
print(images)
```

```python
i = 1
for image in images:
    image = cv2.imread(image, 1)
    plt.imshow(image[1650:2350,1350:2100])
    plt.savefig(f"evolving_map_{i}.png", dpi=600)
    i = i + 1
```
